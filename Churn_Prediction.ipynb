```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
## Load the h5 and pkl files
model = load_model("Models/Churn_Classifier/model.h5")
with open("Pickles/Churn_Classifier/StandardScaler.pkl","rb") as f:
    scaler = pickle.load(f)
with open("Pickles/Churn_Classifier/Geo_encoder.pkl","rb") as f:
    Geo_encoder = pickle.load(f)
with open("Pickles/Churn_Classifier/label_encoder_gender.pkl","rb") as f:
    label_encoder= pickle.load(f)
```

```python
input_data={
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
Geo_encoded = Geo_encoder.transform([[input_data['Geography']]]).toarray()
Geo_encoded_df = pd.DataFrame(Geo_encoded,columns = Geo_encoder.get_feature_names_out(["Geography"]))
Geo_encoded_df
```

```python
label_encoded = label_encoder.transform([input_data['Gender']])
input_data['Gender'] = label_encoded[0]
```

```python
input = pd.DataFrame([input_data])
input.drop(['Geography'],axis=1,inplace = True)
input
```

```python
input = pd.concat([input,Geo_encoded_df],axis=1)
```

```python
input = scaler.transform(input)
input
```

```python
prediction = model.predict(input)[0][0]
prediction
```

```python
if prediction > 0.5:
    print("The customer is likely to churn.")
else:
    print("The customer is unlikely to churn.")
```
